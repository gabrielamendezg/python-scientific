Vega-Lite es una gramática en JSON descriptiva para representar visualizaciones.

Altair por otro lado, es una herramienta de visualización.  Se pueden combinar de manera directa.

# ALTAIR
https://altair-viz.github.io/

```python
import pandas as pd
import altair as alt

# from vega_datasets import data

url = "https://raw.githubusercontent.com/aaizemberg/vis/gh-pages/narda/df_NardaLepes.tsv"
df = pd.read_csv(url, sep="\t" )

alt.Chart(df).mark_circle(size=60).encode(
    x='fav',
    y='retweet',
    tooltip=['text']
).interactive()
```
